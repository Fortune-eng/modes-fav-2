```
# prompt: image segmentation(region based segmentation, edge detection, otsu's threstholding, k-means clustering, morphological operations ) using images in this path /content/malaria-plasmodium-vivax-specie/images

import cv2
import numpy as np
from matplotlib import pyplot as plt
import os

image_dir = "/content/malaria-plasmodium-vivax-specie/images"

def segment_image(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Could not read image at {image_path}")
        return

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    #1. Otsu's Thresholding
    ret, thresh_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # 2. Edge Detection (Canny)
    edges = cv2.Canny(gray, 50, 150) # Adjust thresholds as needed

    #3. Morphological Operations (optional, for refining the segmentation)
    kernel = np.ones((5, 5), np.uint8)  # Adjust kernel size as needed
    opening = cv2.morphologyEx(thresh_otsu, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)


    # Display the results (optional)
    plt.figure(figsize=(12, 6))
    plt.subplot(231), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
    plt.subplot(232), plt.imshow(gray, cmap='gray'), plt.title('Grayscale')
    plt.subplot(233), plt.imshow(thresh_otsu, cmap='gray'), plt.title("Otsu's Thresholding")
    plt.subplot(234), plt.imshow(edges, cmap='gray'), plt.title('Edge Detection (Canny)')
    plt.subplot(235), plt.imshow(opening, cmap='gray'), plt.title('Opening')
    plt.subplot(236), plt.imshow(closing, cmap='gray'), plt.title('Closing')
    plt.show()

# Process the first 5 images in the directory
for i, filename in enumerate(os.listdir(image_dir)):
    if filename.endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(image_dir, filename)
        segment_image(image_path)
        if i >= 4:
          break
```

```
# prompt: feature extraction (GLCM, HOG, SIFT, PCA, and LBP),  using images in this path /content/malaria-plasmodium-vivax-specie/images.

!pip install scikit-image mahotas

from skimage.feature import graycomatrix, graycoprops, local_binary_pattern
from skimage import io
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import cv2
import numpy as np
from skimage.filters import sobel
from skimage.transform import resize
from skimage.feature import hog
import mahotas

def extract_glcm_features(image):
    image = rgb2gray(image)
    glcm = graycomatrix(image.astype(np.uint8), distances=[1], angles=[0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256, symmetric=True, normed=True)
    features = []
    for prop in ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']:
      features.append(graycoprops(glcm, prop).ravel())
    return np.hstack(features)


def extract_hog_features(image):
    resized_img = resize(image, (128, 64))
    fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, channel_axis=-1)
    return fd

def extract_lbp_features(image):
    gray = rgb2gray(image)
    radius = 3
    n_points = 8 * radius
    lbp = local_binary_pattern(gray, n_points, radius, method='uniform')
    hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-7)
    return hist

def extract_sift_features(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sift = cv2.SIFT_create()
    kp, des = sift.detectAndCompute(gray, None)
    return des


def extract_haralick_features(image):
    image = rgb2gray(image)
    textures = mahotas.features.haralick(image.astype(np.uint8))
    ht_mean = textures.mean(axis=0)
    return ht_mean


image_dir = "/content/malaria-plasmodium-vivax-specie/images"

for filename in os.listdir(image_dir):
    if filename.endswith((".png", ".jpg", ".jpeg")):
        image_path = os.path.join(image_dir, filename)
        try:
          image = io.imread(image_path)
          glcm_features = extract_glcm_features(image)
          hog_features = extract_hog_features(image)
          lbp_features = extract_lbp_features(image)
          sift_features = extract_sift_features(image)
          haralick_features = extract_haralick_features(image)

          print(f"Processed image: {filename}")
          print("GLCM Features:", glcm_features)
          print("HOG Features:", hog_features)
          print("LBP Features:", lbp_features)
          print("SIFT Features:", sift_features)
          print("Haralick Features:", haralick_features)

        except Exception as e:
          print(f"Error processing image {filename}: {e}")
```

```
# === CNN-based Image Classification for Malaria Images ===

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Image parameters
IMG_WIDTH = 128
IMG_HEIGHT = 128
image_size = (IMG_WIDTH, IMG_HEIGHT)
BATCH_SIZE = 64
BUFFER_SIZE = tf.data.AUTOTUNE
NUM_CLASSES = 10  # Adjust this if you know actual number of classes

# Data directory
data_dir = "/content/malaria-plasmodium-vivax-specie/images"

# === Image Preprocessing Function ===
def preprocess_image(image_path):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_png(image, channels=3)
    image = tf.image.resize(image, image_size)
    image = tf.image.convert_image_dtype(image, tf.float32)
    return image

# === Load Image Paths ===
image_paths = [os.path.join(data_dir, filename)
               for filename in os.listdir(data_dir)
               if filename.endswith((".png", ".jpg", ".jpeg"))]

num_samples = len(image_paths)

# === MOCK LABELS (Replace with real labels later) ===
labels = np.random.randint(0, NUM_CLASSES, num_samples)
labels_categorical = keras.utils.to_categorical(labels, num_classes=NUM_CLASSES)

# === Create tf.data.Dataset ===
def load_and_preprocess(image_path, label):
    image = preprocess_image(image_path)
    return image, label

image_label_dataset = tf.data.Dataset.from_tensor_slices((image_paths, labels_categorical))
image_dataset = image_label_dataset.map(load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
image_dataset = image_dataset.batch(BATCH_SIZE).prefetch(BUFFER_SIZE)

# === CNN Model ===
model = keras.Sequential([
    keras.Input(shape=(IMG_WIDTH, IMG_HEIGHT, 3)),
    layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Flatten(),
    layers.Dropout(0.5),
    layers.Dense(NUM_CLASSES, activation="softmax")
])

# Compile the model
model.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)

model.summary()

# === Train the Model ===
history = model.fit(
    image_dataset,
    epochs=10,
    steps_per_epoch=num_samples // BATCH_SIZE
)

# === Plot Accuracy ===
plt.plot(history.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.grid(True)
plt.show()

# === Get Predictions for Confusion Matrix ===
# Create a new dataset without batching for prediction
unbatched_dataset = image_label_dataset.map(load_and_preprocess)

# Convert dataset to NumPy for prediction
images_np = np.array([img.numpy() for img, _ in unbatched_dataset])
labels_np = np.array([np.argmax(label.numpy()) for _, label in unbatched_dataset])

# Predict class probabilities, then class indices
pred_probs = model.predict(images_np, batch_size=BATCH_SIZE)
y_pred = np.argmax(pred_probs, axis=1)
y_true = labels_np

# === Confusion Matrix ===
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(10, 10))
im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
ax.figure.colorbar(im, ax=ax)

ax.set(xticks=np.arange(NUM_CLASSES),
       yticks=np.arange(NUM_CLASSES),
       xticklabels=np.arange(NUM_CLASSES),
       yticklabels=np.arange(NUM_CLASSES),
       title='Confusion Matrix',
       ylabel='True Label',
       xlabel='Predicted Label')

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Annotate cells
fmt = 'd'
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, format(cm[i, j], fmt),
                ha="center", va="center",
                color="white" if cm[i, j] > thresh else "black")

fig.tight_layout()
plt.show()
```
